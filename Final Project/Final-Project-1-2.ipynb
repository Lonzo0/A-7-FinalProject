```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import csv
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
#Webscrape Gamma variant Data 
baseurl_Gamma = 'https://www.statista.com/statistics/1246259/number-gamma-variant-worldwide-by-country/?fbclid=IwAR38QwRsTOL3dz72-yh05cZyAiOc7Uy4ABJkzJrpBSvC4t6z8sMDQ2pGtbQ'
headers_Gamma = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.102 Safari/537.36'}
response_Gamma = requests.get(baseurl_Gamma, headers = headers_Gamma).content
soup_Gamma = BeautifulSoup(response_Gamma, 'html.parser')
datahtml_Gamma = soup_Gamma.select('#statTableHTML td')
data_Gamma = []
for data1_Gamma, data2_Gamma in zip(datahtml_Gamma[::2], datahtml_Gamma[1::2]):
    data_Gamma.append({'Country':data1_Gamma.text, 'Number of Cases': data2_Gamma.text})

#Dataframe for Gamma Variant
df_Gamma = pd.DataFrame(data_Gamma)
df_Gamma.to_csv("./output/GammaVariant.csv", index = False)
```

```python
#Read Gamma Variant File 
filename_Gamma = './output/GammaVariant.csv'
dataset_Gamma = []
with open(filename_Gamma,'r') as f_Gamma:
    csvreader_Gamma = csv.DictReader(f_Gamma)
    dataset_Gamma = [r for r in csvreader_Gamma]

#For loop to get number of Gamma cases and country 
dataset_country = []
dataset_number_of_cases=[]
for i in dataset_Gamma:
    cases = i['Number of Cases'] 
    country = i['Country']
    dataset_country.append(country)
    dataset_number_of_cases.append(cases)
dataset_country.reverse()
dataset_number_of_cases.reverse()
dataset_country10 = dataset_country[-10:]
dataset_number = dataset_number_of_cases[-10:]

#For loop to remove comma
#The number of cases in the list are strings. Thus, removing the comma turned them to integers.
y_Gamma = []
for i in dataset_number:
    x = int(i.replace(',',""))
    y_Gamma.append(x)

#Vertical Bar Graph for Gamma Variant
import matplotlib.pyplot as plt
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha = 'center')

#Adjusting the size 
plt.figure(figsize = (10, 5))
plt.xticks(rotation =45)

#Adding labels 
plt.bar(dataset_country10, y_Gamma)
addlabels(dataset_country10, y_Gamma)
plt.title(" Top 10 Countries in terms of Number Of Gamma Variant Cases")
plt.xlabel("Countries")
plt.ylabel("Number of Gamma Variant Cases")
plt.savefig('./output2/GammaCases.png')

#print the vertical bar graph
plt.show()
```

```python
#Total Number of Cases per country and also the variables needed for the following cells

#Read world variant file
filename_World = './output/WorldVariant.csv'
dataset_World = []
with open(filename_World,'r') as f_World:
    csvreader_World = csv.DictReader(f_World)
    dataset_World = [r for r in csvreader_World]

#For loop to get number of  covid cases corresponding to the top 6 country with the most Covid cases 
dataset_country = []
dataset_number_of_cases =[]
for i in dataset_World:
    cases = i['Number of Cases'] 
    country = i['Country']
    dataset_country.append(country)
    dataset_number_of_cases.append(cases)
dataset_country.reverse()
dataset_number_of_cases.reverse()
dataset_country10 = dataset_country[-6:]
dataset_number = dataset_number_of_cases[-6:]

#For loop to remove comma
#The number of cases in the list are strings. Thus, removing the comma turned them to integers.
y_World = []
for i in dataset_number:
    x = int(i.replace(',',""))
    y_World.append(x)
y_World.reverse()

#Assign Variables for each country 
USA_total = y_World[0]
India_total = y_World[1]
Brazil_Total = y_World[2]
Russia_Total = y_World[3]
France_Total = y_World[4]
UK_Total = y_World[5]

#Creating a List of covid variants that corresponds to the order of the number of cases
label = ['Gamma','Beta','Alpha','Delta']
variants = ['Gamma','Beta','Alpha','Delta','Original and Other']
```

```python
#Pie Chart For the different variants in the USA

#Read USA file
filename_USA = "./output/USA.csv"
dataset_USA = []
with open(filename_USA,'r') as f_USA:
    csvreader_USA = csv.DictReader(f_USA)
    dataset_USA = [r for r in csvreader_USA]
    
#for loop to get the number of cases of USA per variant 
dataset_number_of_cases =[]
for i in dataset_USA:
    cases = i['Number of Cases'] 
    dataset_number_of_cases.append(cases)
dataset_number = dataset_number_of_cases[-10:]

#For loop to remove comma
#The number of cases in the list are strings. Thus, removing the comma turned them to integers.
y_USA = []
for i in dataset_number:
    x = int(i.replace(',',""))
    y_USA.append(x)
    
#to get the total number of cases among the 4 variants in USA
total = sum(y_USA)

#to get the number of the other covid variants in USA
USA_Other_Variants = (USA_total - total)

#Pie Chart 
plt.pie(y_USA, labels = label, autopct ='%0.2f%%', shadow = True, explode = [0,0,0.3,0])
plt.savefig('./output2/USACases.png')
plt.title('The Percentage of the Different Variants in the US')
plt.axis("equal")

#print pie chart 
plt.show()

```

```python
#Bar Graph as comparison for the difference in amount of the different variants in the USA
total = sum(y_USA)
USA_Other_Variants = (USA_total - total)
y_USA.append(USA_Other_Variants)

#Create the Vertical Bar Graph 
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha = 'center')

#Adjust the size
plt.figure(figsize = (10, 5))
plt.xticks(rotation =45)

#add labels
plt.bar(variants, y_USA)
addlabels(variants, y_USA)
plt.title("Number of Cases per Variant in the US")
plt.xlabel("Type of Variant")
plt.savefig('./output2/VariantUSACases.png')
plt.ylabel("Number of Covid Cases")

#print the graph
plt.show()
```
